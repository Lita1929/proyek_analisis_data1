#### Eksplorasi Eksplorasi Data customers_df & orders_df

Mengidentifikasi pelanggan yang belum pernah melakukan order. Untuk melakukan hal ini, kita bisa membuat sebuah kolom baru bernama “status” pada data customers_df. Kolom tersebut memiliki nilai “Active” untuk pelanggan yang pernah melakukan order setidaknya sekali dan sebaliknya bernilai “Non Active” untuk pelanggan yang belum pernah melakukan order sama sekali.

```python
customer_id_in_orders_df =  orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")

customers_df.sample(5)
```

```python
customers_df.groupby(by="status").customer_id.count()
```

Seperti yang terlihat pada pivot table di atas, ternyata semua pelanggan pernah melakukan transaksi sebelumnya.

```python
orders_df.describe(include="all")
```

```python
order_customer_df = pd.merge(
   left=orders_df,
   right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
order_customer_df.head()
```

### Pertanyaan 4 - 6:

**RFM Analysis**

- Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.- 
Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi
- 
Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
rfm_df = all_data_df.groupby(by=["customer_id", "customer_num"], as_index=False).agg({
    "order_purchase_timestamp": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "payment_value": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["customer_id", "customer_num","max_order_timestamp", "frequency", "monetary"]
 
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
 
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

Mengurutkan data berdasarkan recency, frequency dan monetary. Untuk diambil nilai dari customer_num

```python
recency_sorted = rfm_df.sort_values(by='recency', ascending=True)
frequency_sorted = rfm_df.sort_values(by='frequency', ascending=False)
monetary_sorted = rfm_df.sort_values(by='monetary', ascending=False)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
jumlah_kategori = 5
colors = ["#AC8BEE"] * (jumlah_kategori)
 
sns.barplot(y="recency", x="customer_id", hue="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0], legend=False)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
# Mengatur ticks untuk sumbu x
ax[0].set_xticks(range(len(rfm_df.head(5))))
# Mengambil nilai customer_num
ax[0].set_xticklabels(recency_sorted['customer_num'].head(5).values)
 
sns.barplot(y="frequency", x="customer_id", hue="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1], legend=False)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].set_xticks(range(len(rfm_df.head(5))))
ax[1].set_xticklabels(frequency_sorted['customer_num'].head(5).values)
 
sns.barplot(y="monetary", x="customer_id", hue="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2], legend=False)
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
ax[2].set_xticks(range(len(rfm_df.head(5))))
ax[2].set_xticklabels(monetary_sorted['customer_num'].head(5).values)
 
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=25)
plt.show()
```
